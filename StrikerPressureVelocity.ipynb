```python
import pandas as pd
import matplotlib.pyplot as plt 
import os


data = pd.read_csv("../config/Bar_Wave_Speed.csv")
```

```python
data["Striker Pressure"] = data["Striker Pressure"]* 145.038 # converts from MPa to psi 
```

```python
plt.figure(figsize = (10,6))
plt.scatter(data["Striker Pressure"], data["Striker Velocity"], s = 25, c="royalblue")
plt.title("Striker Velocity vs Pressure (12in Striker)")
plt.xlabel("Velocity (m/s)", fontsize = 14)
plt.ylabel("Striker Pressure (psi)", fontsize = 14)
plt.grid(alpha=0.3)
plt.savefig("../config/StikerVelocityvsPressure.png", dpi=300)
plt.show()
```
